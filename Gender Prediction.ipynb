# Gender Prediction Using Height and Weight

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
df = pd.read_csv('weight-height.csv')
```

```python
df
```

```python
pd.get_dummies(df.Gender)
```

```python
dummies = pd.get_dummies(df.Gender)
```

```python
pd.concat([df,dummies],axis='columns')
```

```python
merged=pd.concat([df,dummies],axis='columns')
```

```python
merged.drop(['Gender','Male'],axis='columns',inplace=True)
```

```python
print(merged.head())
```

```python
Reg = linear_model.LinearRegression()
```

```python
X = merged.drop('Female',axis='columns')
```

```python
y=merged.Female
```

```python
Reg.fit(X,y)
```

```python
Reg.predict([[74,110]])
```
