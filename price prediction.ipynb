# Predicting Salary Using Dummy Variables and One Hot encoder

```python
import pandas as pd
```

```python
df = pd.read_csv('carprices.csv')
```

```python
df
```

```python
pd.get_dummies(df['Car Model'])
```

```python
dummy=pd.get_dummies(df['Car Model']).astype(int)
```

```python
dummy
```

```python
pd.concat([df,dummy],axis='columns')
```

```python
final = pd.concat([df,dummy],axis='columns')
```

```python
final.drop(['Car Model','Mercedez Benz C class'],axis='columns')
```

```python
fdf=final.drop(['Car Model','Mercedez Benz C class'],axis='columns')
```

```python
X = fdf.drop(['Sell Price($)'],axis='columns')
```

```python
X
```

```python
y = fdf['Sell Price($)']
```

```python
y
```

```python
from sklearn import linear_model
```

```python
Model = linear_model.LinearRegression()
```

```python
Model.fit(X,y)
```

```python
Model.predict([[70000,5,0,0]])
```

```python
df
```

```python
dfp=pd.read_csv('to predict.csv')
```

```python
dfp
```

```python
Model.predict(dfp)
```

```python
pvalues = Model.predict(dfp)
```

```python
pvalues
```

```python
dfp['Predicted Price'] = pvalues
```

```python
dfp.to_csv('PredictedPrice.csv')
```
